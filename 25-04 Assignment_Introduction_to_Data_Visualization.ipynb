# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

### Expected output:


```python
# Your code here

import matplotlib.pyplot as plt
import numpy as np

# Sample monthly sales data
months = np.arange(1, 13)  
facecream_sales = [2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900]
facewash_sales = [1500, 1200, 1340, 1130, 1750, 1555, 1120, 1400, 1780, 1890, 2100, 1760]
bar_width = 0.4 

# Bar positions
bar1 = months - bar_width / 2
bar2 = months + bar_width / 2

# Plotting the bars
plt.bar(bar1, facewash_sales, width=bar_width, label='facewash', color='dodgerblue')
plt.bar(bar2, facecream_sales, width=bar_width, label='facecream', color='darkorange')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('Face Cream and Facewash Product Sales Month Wise')
plt.xticks(months)  # Set x-axis ticks to month numbers
plt.legend()

# Show plot
plt.show()
```

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

### Expected Output:



```python
# Your code here
import matplotlib.pyplot as plt
import numpy as np

# Sample data
months = np.arange(1, 13)
facecream = [2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540,1990,2340,2900]
facewash = [1500, 1200,1340,1130,1740,1555,1120,1400,1780,1890, 2100,1760]
toothpaste = [5200, 5100,4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400]
bathingsoap = [9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400]
shampoo = [1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800]
moisturizer = [1500, 1200,1340, 1130,1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760]

# Plotting each product line
plt.plot(months, facecream, label='facecream', color='red',marker='o')
plt.plot(months, facewash, label='facewash', color='blue',marker='o')
plt.plot(months, toothpaste, label='toothpaste', color='green',marker='o', linestyle='--')
plt.plot(months, bathingsoap, label='bathingsoap', color='yellow',marker='o')
plt.plot(months, shampoo, label='shampoo', color='black',marker='o')
plt.plot(months, moisturizer, label='moisturizer', color='skyblue',marker='o')

# Add labels, legend, and title
plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('Products Sales Month Wise')
plt.legend()

# Show the plot
plt.show()



```
